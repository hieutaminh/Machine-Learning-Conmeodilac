###  Tiền xử lý dữ liệu đưa dữ liệu từ dạng text về dạng ma trận

- Thử nghiệm để kiểm tra hoạt động chuyển hoá dữ liệu về dạng ma trận

```python
# load dữ liệu các stopwords
with open("/content/drive/MyDrive/Colab Notebooks/ML-exercises/Ex7/data/vietnamese-stopwords.txt",encoding="utf-8") as f:
    stopwords = f.readlines()
stopwords = [x.strip().replace(" ", "_") for x in stopwords]
print("Danh sách 10 từ dừng đầu tiên (từ không mang ý nghĩa phân loại): ", stopwords[:10])

# Transforming data
# Chuyển hoá dữ liệu text về dạng vector tfidf
#     - loại bỏ từ dừng
#     - sinh từ điển
module_count_vector = CountVectorizer(stop_words=stopwords)
model_rf_preprocess = Pipeline([('vect', module_count_vector),
                    ('tfidf', TfidfTransformer()),
                    ])
# Hàm thực hiện chuyển đổi dữ liệu text thành dữ liệu số dạng ma trận
# Input: Dữ liệu 2 chiều dạng numpy.array, mảng nhãn id dạng numpy.array
data_preprocessed = model_rf_preprocess.fit_transform(data_train.data, data_train.target)
print("10 từ đầu tiên trong từ điển:")
i = 0
for k,v in module_count_vector.vocabulary_.items():
    i+=1
    print(i, ": ", (k, v))
    if i > 10:
        break
```

### Bài 1: Sử dụng trực tiếp TfidfVectorizer()
https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html

```python
import pandas as pd

# 1. Thực hiện biến đổi dữ liệu về dạng Tf-Idf trực tiếp sử dụng TfidfVectorizer()
# Khởi tạo TfidfVectorizer với stopwords đã được tải
vectorizer = TfidfVectorizer(stop_words=stopwords)

# Sử dụng toàn bộ dữ liệu hoặc X_train nếu đã chạy ô chia dữ liệu
tfidf_matrix = vectorizer.fit_transform(data_train.data)

feature_names = vectorizer.get_feature_names_out()

# Lấy vector của văn bản đầu tiên
doc1_tfidf = tfidf_matrix[0].toarray()[0]

df = pd.DataFrame({
    'word': feature_names,
    'tfidf': doc1_tfidf
})

# Hiển thị 10 từ có giá trị cao nhất
df = df.sort_values(by='tfidf', ascending=False).head(10)

print(df)
```

### Chia dữ liệu làm 2 phần training và testing

- Training chiếm 80 % dữ liệu
- Testing chiếm 20 % dữ liệu

```python
from sklearn.model_selection import ShuffleSplit

# chia dữ liệu thành 2 phần sử dụng hàm train_test_split.
test_size = 0.2
# cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
X_train, X_test, y_train, y_test = train_test_split( data_preprocessed, data_train.target, test_size=test_size)

# hiển thị một số thông tin về dữ liệu
print("Dữ liệu training: ", X_train.shape, y_train.shape)
print("Dữ liệu testing: ", X_test.shape, y_test.shape)
print("Danh sách nhãn và id tương ứng: ", [(idx, name) for idx, name in enumerate(data_train.target_names)] )
```

## 2. Training svm model

Sử dụng thư viện sklearn để xây dựng mô hình
-  Thử trước với hàm nhân phân tách là Linear, tham số C=1.0
- `svm.SVC(kernel='linear', C=1.0)`


```python
print("- Training ...")
print("- Train size = {}".format(X_train.shape))
model = svm.SVC(kernel='linear', C=1)
model.fit(X_train, y_train)
print("- model - train complete")
```

## 3. Testing svm model

Thực hiện dự đoán nhãn cho từng văn bản trong tập test

Độ đo đánh giá:
> accuracy = tổng số văn bản dự đoán đúng  / tổng số văn bản có trong tập test  

```python
from sklearn.metrics import accuracy_score
print("- Testing ...")
y_test_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
print("- Test Acc = {}".format(accuracy_score(y_test, y_test_pred)))
print("- Train Acc = {}".format(accuracy_score(y_train, y_train_pred)))
```

### Bài 2: thực hiện lại các bước trên với kernel = 'rbf'

```python
# code
model2 = svm.SVC(kernel='rbf', C=1)
model2.fit(X_train, y_train)
print("- model rbf - train complete")
y_test_pred = model2.predict(X_test)
y_train_pred = model2.predict(X_train)
print("- Test Acc = {}".format(accuracy_score(y_test, y_test_pred)))
print("- Train Acc = {}".format(accuracy_score(y_train, y_train_pred)))
```

### 3.1. Thực hiện sử dụng model đã được train để infer 1 văn bản mới
- Dữ liệu mới đến ở dạng dữ liệu thô => cần tiền xử lý dữ liệu về dạng dữ_liệu_ma_trận
- infer sử dụng hàm model.predict(dữ_liệu_ma_trận)

```python
# Tiền xử lý dữ liệu sử dụng module model_rf_preprocess.
new_doc = "Công phượng ghi bàn cho đội tuyển Việt nam"
# Trước hết, cần thực hiện tách từ sử dụng pyvi
tokenized_new_doc = ViTokenizer.tokenize(new_doc)
# Cần đưa văn bản ở dạng mảng/vector
tokenized_new_doc = [tokenized_new_doc]
print(tokenized_new_doc)
# Rồi sử dụng module model_rf_preprocess
input_data_preprocessed = model_rf_preprocess.transform(tokenized_new_doc)
print(input_data_preprocessed)
```

### Bài 3: dự đoán nhãn của văn bản trên

```python
# code
print("Dự đoán nhãn cho văn bản mới bằng linear: ", data_train.target_names[model.predict(input_data_preprocessed)[0]])
print("Dự đoán nhãn cho văn bản mới bằng rbf: ", data_train.target_names[model2.predict(input_data_preprocessed)[0]])
```

## 4. Bài tập bổ sung:

### 4.1 Thử nghiệm các tham số

- Các tham số với giá trị khác nhau có thể ảnh hưởng để kết quả học
- Cần thử nghiệm kỹ lượng để đưa ra kết quả khách quan: tham số C, gamma, kernel.
    - Chọn mô hình với bộ tham số cho kết quả tốt nhất
- Gợi ý:
    - https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html
    - Sử dụng grid search

#### Bài 4: Vẽ Learning curve khảo sát Acc của SVM-linear với tham số C thay đổi

```python
# code
# tham khảo miền của C
list_C = [0.001, 0.01, 0.1, 1, 5.0, 10.0, 100]
train_accs = []
test_accs = []

for c in list_C:
    # Khởi tạo và huấn luyện model với từng giá trị C
    clf = svm.SVC(kernel='linear', C=c)
    clf.fit(X_train, y_train)

    # Tính accuracy trên tập train và tập test
    train_acc = accuracy_score(y_train, clf.predict(X_train))
    test_acc = accuracy_score(y_test, clf.predict(X_test))

    train_accs.append(train_acc)
    test_accs.append(test_acc)
    print(f"C = {c}: Train Acc = {train_acc:.4f}, Test Acc = {test_acc:.4f}")

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.plot(list_C, train_accs, marker='o', label='Train Accuracy')
plt.plot(list_C, test_accs, marker='s', label='Test Accuracy')
plt.xscale('log')
plt.xlabel('Tham số C (log scale)')
plt.ylabel('Accuracy')
plt.title('Learning Curve của SVM Linear theo tham số C')
plt.legend()
plt.grid(True)
plt.show()
```

#### Bài 5: Sử dụng GridSearchCV để tìm bộ tham số tốt nhất

```python
# code
# Có thể tham khảo giá trị các tham số như sau
params_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100],
          'gamma': [0.0001, 0.001, 0.01, 0.1],
          'kernel':['linear','rbf', 'poly'] }

# Khởi tạo GridSearchCV
grid_search = GridSearchCV(svm.SVC(), params_grid, cv=5, n_jobs=-1, verbose=1)

# Thực hiện tìm kiếm trên tập huấn luyện
print("Đang tìm kiếm bộ tham số tối ưu...")
grid_search.fit(X_train, y_train)

# Hiển thị kết quả tốt nhất
print("Bộ tham số tốt nhất: ", grid_search.best_params_)
print("Độ chính xác cao nhất trên tập validation: ", grid_search.best_score_)

# Đánh giá trên tập test với model tốt nhất
best_model = grid_search.best_estimator_
test_acc = accuracy_score(y_test, best_model.predict(X_test))
print("Accuracy của model tốt nhất trên tập test: ", test_acc)
```

### 4.2 Phân loại số viết tay

```python
# Standard scientific Python imports
import matplotlib.pyplot as plt

# Import datasets, classifiers and performance metrics
from sklearn import datasets, svm, metrics

# The digits dataset
digits = datasets.load_digits()

# The data that we are interested in is made of 8x8 images of digits, let's
# have a look at the first 4 images, stored in the `images` attribute of the
# dataset.  If we were working from image files, we could load them using
# matplotlib.pyplot.imread.  Note that each image must have the same size. For these
# images, we know which digit they represent: it is given in the 'target' of
# the dataset.
images_and_labels = list(zip(digits.images, digits.target))
for index, (image, label) in enumerate(images_and_labels[:4]):
    plt.subplot(2, 4, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Training: %i' % label)

# To apply a classifier on this data, we need to flatten the image, to
# turn the data in a (samples, feature) matrix:
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
target = digits.target
X_train, X_test, y_train, y_test = train_test_split( data, target, test_size=test_size)

print("Dữ liệu training = ", X_train.shape, y_train.shape)
print("Dữ liệu testing = ", X_test.shape, y_test.shape)
```

#### Bài 6: phân loại với dữ liệu trên

```python
# 1. Khởi tạo và huấn luyện model SVM cho bài toán phân loại chữ số
# Sử dụng GridSearchCV để tìm tham số tốt nhất cho tập dữ liệu digits
params_digits = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.001, 0.01, 0.1, 1],
    'kernel': ['rbf', 'poly']
}

print("- Đang tìm kiếm tham số tối ưu cho dữ liệu Digits...")
grid_digits = GridSearchCV(svm.SVC(), params_digits, cv=5)
grid_digits.fit(X_train, y_train)

print("Bộ tham số tốt nhất:", grid_digits.best_params_)

# 2. Đánh giá trên tập test
best_svc_digits = grid_digits.best_estimator_
y_pred = best_svc_digits.predict(X_test)

print(f"Accuracy trên tập test: {metrics.accuracy_score(y_test, y_pred):.4f}")

# 3. Hiển thị báo cáo chi tiết
print("\nClassification Report:\n", metrics.classification_report(y_test, y_pred))

# Hiển thị một số kết quả dự đoán
images_and_predictions = list(zip(X_test, y_pred))
plt.figure(figsize=(10, 4))
for index, (image, prediction) in enumerate(images_and_predictions[:4]):
    plt.subplot(1, 4, index + 1)
    plt.axis('off')
    plt.imshow(image.reshape(8, 8), cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Pred: %i' % prediction)
plt.show()
```
