2. Cho đoạn code minh họa Biểu đồ thể hiện số bài đăng theo tháng của năm 2020. Hãy thực hiện các yêu cầu cho các câu hỏi dưới đây:

```python
df1 = newdf.groupby('month').size().reset_index(name='number of records')
n = df1['month'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df1['month'], df1['number of records'], color=c, width=.5)
for i, val in enumerate(df1['number of records'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticks(range(len(df1['month'])))
plt.gca().set_xticklabels(df1['month'], rotation=60, horizontalalignment= 'right')
plt.title("Biểu đồ thể hiện số bài đăng theo tháng của năm 2020", fontsize=22)
plt.ylabel('number of records')
plt.show()
```

* Biểu đồ thể hiện giá nhà đất trung bình theo tháng (VND/m2)

```python
# Tính toán cột mới: giá trên mỗi m2
df['price_per_unit'] = df['price'] / df['square']

# Tính trung bình cộng cột 'price_per_unit' theo từng tháng
df2 = df.groupby('month')['price_per_unit'].mean().reset_index(name='avg_price')

n = df2['month'].unique().__len__() + 1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(67)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi = 80)
plt.bar(df2['month'], df2['avg_price'], color=c, width = 0.5)
for i, val in enumerate(df2['avg_price'].values):
    plt.text(i, val, round(float(val), 2), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight' : 500, 'size':12})

plt.gca().set_xticks(range(len(df2['month'])))
plt.gca().set_xticklabels(df2['month'], rotation=60, horizontalalignment= 'right')
plt.title("Biểu đồ thể hiện giá nhà đất trung bình theo tháng (VND/m2)", fontsize=22)
plt.ylabel('Giá trung bình')
plt.show()
```

* Biểu đồ mật độ giá nhà của tháng 4 và tháng 12 (VND/m2)

```python
plt.figure(figsize=(13,10), dpi= 80)

# Sử dụng 'df' thay vì 'dftmp' và cột 'price_per_unit' vừa tạo
sns.histplot(df.loc[df['month'] == '12', "price_per_unit"], color="dodgerblue", label="Tháng 12",kde=True, stat="density", alpha=0.7)
sns.histplot(df.loc[df['month'] == '04', "price_per_unit"], color="orange", label="Tháng 4", kde=True, stat="density", alpha=0.7)

plt.xlim(0, 150)
plt.title('Biểu đồ mật độ giá nhà của tháng 4 và tháng 12 (VND/m2)', fontsize=22)
plt.xlabel('Giá (VND/m2)')
plt.legend()
plt.show()
```

```python
# Lọc dữ liệu cho tháng 4 và tháng 12
df_month_4 = df[df['month'] == '04']['price']
df_month_12 = df[df['month'] == '12']['price']

# Vẽ biểu đồ mật độ
plt.figure(figsize=(12, 6))
sns.kdeplot(df_month_4, fill=True, label='Tháng 4', color="g")
sns.kdeplot(df_month_12, fill=True, label='Tháng 12', color="b")

plt.title('Biểu đồ mật độ giá nhà của tháng 4 và tháng 12', fontsize=18)
plt.xlabel('Giá (triệu VND/m2)')
plt.ylabel('Mật độ')
plt.legend()
plt.show()
```

* Số lượng bất động sản phân theo số lượng bedrooms

```python
!pip install squarify
import squarify
df = newdf.groupby('bedrooms').size().reset_index(name='counts')
labels = df.apply(lambda x: str(x[0]) + "\n (" + str(x[1]) + ")", axis=1)
sizes = df['counts'].values.tolist()
colors = [plt.cm.Spectral(i/float(len(labels))) for i in range(len(labels))]

plt.figure(figsize=(12,8), dpi= 80)
squarify.plot(sizes=sizes, label=labels, color=colors, alpha=.8)

plt.title('Số lượng bất động sản phân theo số lượng bedrooms')
plt.axis('off')
plt.show()
```

* Giá bất động sản trung bình theo số lượng bedrooms

```python
df_bed = newdf.groupby('bedrooms')['price_per_unit'].mean().reset_index(name='avg_price')

n = len(df_bed['bedrooms'].unique()) + 1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16, 10), dpi=80)
plt.bar(df_bed['bedrooms'], df_bed['avg_price'], color=c, width=0.3)

for i, val in enumerate(df_bed['avg_price'].values):
    plt.text(df_bed['bedrooms'].iloc[i], val, round(float(val), 2),
             horizontalalignment='center',
             verticalalignment='bottom',
             fontdict={'fontweight': 500, 'size': 12})

# ✅ Sửa ở đây
plt.xticks(df_bed['bedrooms'], df_bed['bedrooms'], rotation=60, ha='right')

plt.title("Giá bất động sản trung bình theo số lượng bedrooms", fontsize=22)
plt.ylabel('Giá trung bình')
plt.show()
```

3. Tương quan giữa các trường thuộc tính

```python
# Gợi ý: sns.heatmap(cmap='RdYlGn', ...)
plt.figure(figsize=(12,10), dpi= 80)

# Chọn các cột số để tính toán tương quan
numeric_cols = ['square', 'bedrooms', 'bathrooms', 'price']
correlation_matrix = newdf[numeric_cols].corr()
sns.heatmap(correlation_matrix, xticklabels=correlation_matrix.columns, yticklabels=correlation_matrix.columns, cmap='RdYlGn', center=0, annot=True)

plt.title('Hệ số tương quan giữa các trường thuộc tính số', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
